```python
#importing all the necessary libraries for machine learning
import pandas as pd 
import numpy as np 
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#loading the boston data
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
#description of the data
print(boston.DESCR)
```

```python
#organising the data
x=pd.DataFrame(boston.data, columns=boston.feature_names)
x
```

```python
# performing feature scaling on x
standardscaler = preprocessing.StandardScaler()
x_scaled= standardscaler.fit_transform(x)
x_scaled
```

```python
#storing target value in y
y=pd.DataFrame(boston.target, columns=['target'])
print(y)
```

```python
#performing feature scaling on y
y_scaled= standardscaler.fit_transform(y)
y_scaled
```

```python
#splitting the data into train and test set
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.2)
```

```python
#finding the length of training set
len(x_train)
```

```python
#using linear model from sklearn library 
regression = linear_model.LinearRegression()
```

```python
#fitting the training data on the linear regression model
regression.fit(x_train, y_train)
```

```python
#parameters of different features
regression.coef_
```

```python
#predicted values
y_predicted_scaled= regression.predict(x_test)
y_predicted_scaled
```

```python
#calculating the MSE from training set
m_sq_err=mean_squared_error(regression.predict(x_train),y_train)
print(m_sq_err)
```

```python
#calculating the generalisation error from test set
generalisation_err= mean_squared_error(regression.predict(x_test), y_test)
print(generalisation_err)
```

```python
#rescaling back the predicted values
y_predicted= standardscaler.inverse_transform(y_predicted_scaled)
y_predicted
```

```python
#showing target and predicted values of test set
p=pd.DataFrame(list(zip(y_test, y_predicted_scaled)), columns=['target','predicted'])
p.head(102)
```
